```python
citydf = df.groupby('City').sum()
citydf['Total'].plot()
plt.xlabel('City / District')
plt.ylabel('Sales')
plt.grid()
```

```python
df['Month'] = df['Date'].str[0:2]

df['Month'] = df['Month'].str.replace("/","")
df.head(50)
```

```python
df['Month'] = df['Month'].astype(int)
df['Month'].nunique()
```

```python
MonthWise = df.groupby('Month').sum()
MonthWise
MonthWise['Total'].plot()
plt.xlabel('Months')
plt.ylabel('Sales')
plt.grid()

```

```python
MonthWise['Total'].plot(kind='bar')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.title('Total Sales by Month')
plt.grid()
plt.show()
```

```python
MonthWise['Total'].plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Total Sales Distribution by Month')
plt.ylabel('')  # No need to label the y-axis for a pie chart
plt.show()
```
