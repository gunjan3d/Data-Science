```python
### converting the months to integer from string
```

```python
df = df[df['Order Date'].str[0:2] != 'Or']
df['Month'] = df['Order Date'].str[0:2]
df
```

```python
df['Month'] = df['Month'].astype('int32')
df
```

```python
## making a sales column for easy understanding
```

```python
df['Sales'] = df['Quantity Ordered'].astype('float') * df['Price Each'].astype('float')
df['Quantity Ordered'] = pd.to_numeric(df['Quantity Ordered'],errors = 'coerce')
df['Price Each'] = pd.to_numeric(df['Price Each'],errors = 'coerce')
### dont know  why this warning is showing up
```

```python
df.loc[:, 'Sales'] = df['Quantity Ordered'].astype('float') * df['Price Each'].astype('float')
df.loc[:, 'Quantity Ordered'] = pd.to_numeric(df['Quantity Ordered'], errors='coerce')
df.loc[:, 'Price Each'] = pd.to_numeric(df['Price Each'], errors='coerce')
```

```python
### SALES ACCORDING TO THIER RESPECTIVE MONTHS
```

```python
results = df.groupby('Month').sum()
```

```python
import matplotlib.pyplot as plt

months = range(1,13)

plt.bar(months,results['Sales'])

plt.xticks(months)
plt.xlabel("months in Number")
plt.ylabel("sales in 100k USD")
plt.show()
```
