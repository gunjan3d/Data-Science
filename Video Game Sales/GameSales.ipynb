```python
df = df.dropna()
```

```python
df.head()
```

```python
df_genre = df.groupby('Genre').sum()
ax = df_genre['NA_Sales'].plot(kind='bar')
plt.title('Total North American Sales by Genre')
plt.xlabel('Genre')
plt.ylabel('Total Sales (in millions)')
for p in ax.patches:
    ax.annotate(str(round(p.get_height(), 2)), (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, 10), textcoords='offset points')
plt.grid()
plt.show()
```

```python
ax = df_genre['EU_Sales'].plot(kind='bar')
plt.title('Total EU Sales by Genre')
plt.xlabel('Genre')
plt.ylabel('Total Sales (in millions)')
for p in ax.patches:
    ax.annotate(str(round(p.get_height(), 2)), (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, 10), textcoords='offset points')
plt.grid()
plt.show()
```

```python
ax = df_genre['JP_Sales'].plot(kind='bar')
plt.title('Total Japan Sales by Genre')
plt.xlabel('Genre')
plt.ylabel('Total Sales (in millions)')
for p in ax.patches:
    ax.annotate(str(round(p.get_height(), 2)), (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, 10), textcoords='offset points')
plt.grid()
plt.show()
```
